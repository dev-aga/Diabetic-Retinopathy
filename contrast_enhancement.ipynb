# Enhancing whole dataset

```python
folder_path = "./colored_images_mod"

for folder in os.listdir(folder_path):
  for data in os.listdir(folder_path + "/" + folder):
    for image in os.listdir(folder_path + "/" + folder + "/" + data):
      path= folder_path + "/" + folder + "/" + data + "/" + image
      img = cv2.imread(path)
      b,g,r = cv2.split(img)

      clahe = cv2.createCLAHE(clipLimit =2.5, tileGridSize=(8,8))
      g1=clahe.apply(g)

      new_img=cv2.merge([b,g1,r])

      path2= "./new_dataset" + "/" + folder + "/" + data
      if os.path.exists(path2)==False:
        os.makedirs(path2)
      cv2.imwrite(path2 + "/" + image, new_img)
```
